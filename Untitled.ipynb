```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
img = cv2.imread('cat.jpeg',cv2.IMREAD_COLOR)
img = np.array(img, dtype="uint8")
plt.imshow(img)
height, weight = img.shape[:2]
```

```python
def create_watermark(n, m):
    watermark = np.random.randint(2, size = (n , m)).astype("uint8")
    return watermark
```

```python
watermark = create_watermark(height, weight)
print(watermark.shape)
```

```python
plt.imshow(watermark)
```

```python
b, g, r = cv2.split(img)
watermarked_b = cv2.add(b, watermark)
watermarked_g = cv2.add(g, watermark)
watermarked_r = cv2.add(r, watermark)
watermarked = cv2.merge([watermarked_r, watermarked_g, watermarked_b])
plt.imshow(watermarked)
```
